# Geo world population
> Population from excel

Source: pyecharts samples

```python
import pandas as pd
```

```python
EXCEL_URL = 'https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/EXCEL_FILES/1_Population/WPP2019_POP_F01_1_TOTAL_POPULATION_BOTH_SEXES.xlsx'
```

```python
df = pd.read_excel(EXCEL_URL, header=16)
df.head(1)
```

```python
data = df[df['Type']=='Country/Area'][['Region, subregion, country or area *','2020']].values.tolist()
```

```python
high = max([x for _, x in data[1:]])
low = min([x for _, x in data[1:]])
```

```python
from pyecharts.charts import Map
import pyecharts.options as opts

m = (Map().add(
    maptype='world',
    series_name="World Population",
    data_pair=data[1:],
    is_map_symbol_show=False,
    label_opts=opts.LabelOpts(is_show=False),
))
m.render_notebook()
```

```python
m.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(
        min_=low,
        max_=high,
        range_text=["max", "min"],
        is_calculable=True,
        range_color=["lightskyblue", "yellow", "orangered"],
    )
)
m.render_notebook()
```

```python
m = (Map().add(
    maptype='world',
    series_name="World Population",
    data_pair=data[1:],
    is_map_symbol_show=False,
    label_opts=opts.LabelOpts(is_show=False),
    name_map={
        'Dem. Rep. Congo': 'DR Congo',
        'Central African Rep.': 'Central African Republic',
        'S. Sudan': 'South Sudan',
        'Korea': 'South Korea'
    }
).set_global_opts(visualmap_opts=opts.VisualMapOpts(
        min_=low,
        max_=high,
        range_text=["max", "min"],
        is_calculable=True,
        range_color=["lightskyblue", "yellow", "orangered"],
    )
))
m.render_notebook()
```
