# Sankey chart

>Basic sankey charts

```python
import deetly
import pandas as pd

from pyecharts.charts import Sankey
from pyecharts import options as opts
from pyecharts.globals import ChartType
```

```python
# Create datapackage
metadata = {
    "name": "EChart Sankey examples",
    "author": "deetly",
    "keywords": ["line","linechart","chart","examples"]
}

dp = deetly.package(metadata)
```

```python
nodes = [{"name": "category1"}, {"name": "category2"}, {"name": "category3"}]

links = [
    {"source": "category1", "target": "category2", "value": 10},
    {"source": "category2", "target": "category3", "value": 15},
]

g = Sankey().add(
    "sankey",
    nodes,
    links,
    linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
    label_opts=opts.LabelOpts(position="right"),
)

g.render_notebook()
```

```python
dp.echart(g, "Horizontal Sankey sample")
```

```python
nodes = [
    {"name": "a"},
    {"name": "b"},
    {"name": "a1"},
    {"name": "b1"},
    {"name": "c"},
    {"name": "e"},
]
links = [
    {"source": "a", "target": "a1", "value": 5},
    {"source": "e", "target": "b", "value": 3},
    {"source": "a", "target": "b1", "value": 3},
    {"source": "b1", "target": "a1", "value": 1},
    {"source": "b1", "target": "c", "value": 2},
    {"source": "b", "target": "c", "value": 1},
]

g = Sankey().add(
    "sankey",
    nodes,
    links,
    pos_bottom="10%",
    focus_node_adjacency="allEdges",
    orient="vertical",
    levels=[
        opts.SankeyLevelsOpts(
            depth=0,
            itemstyle_opts=opts.ItemStyleOpts(color="#eee"),
            linestyle_opts=opts.LineStyleOpts(color="source", opacity=0.6),
        )
    ],
    linestyle_opt=opts.LineStyleOpts(opacity=0.4, curve=0.5, color="source"),
    label_opts=opts.LabelOpts(position="right"),
)

g.render_notebook()
```

```python
dp.echart(g, "Vertical Sankey sample")
```

```python
dp.publish()
```
