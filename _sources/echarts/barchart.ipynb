# Bar chart

>Basic bar chart

Source: ECharts samples

```python
import deetly
import pandas as pd

from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ChartType
```

```python
# Create datapackage
description = """
ECharts Barchart examples
"""

metadata = {
    "name": "EChart Barchart examples",
    "description":description, 
    "author": "deetly",
    "keywords": ["bar","barchart","examples"],
    "theme":"examples",
    "license": "MIT"
}

dp = deetly.package(metadata)
```

```python
g = (Bar()
    .add_xaxis(["Microsoft", "Amazon", "IBM", "Oracl", "Google", "Alibaba"])
    .add_yaxis('2017-2018 Revenue in (billion $)', [21.2, 20.4, 10.3, 6.08, 4, 2.2])
    .set_global_opts(title_opts=opts.TitleOpts(title="Top cloud providers 2018", subtitle="2017-2018 Revenue"))
)

g.render_notebook()
```

```python
dp.echart(g, "Cloud", "Cloud")
```

```python
dp.publish()
```
