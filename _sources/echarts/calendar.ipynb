# Calendar

>Sample calendar chart

Source: pyecharts test

```python
import deetly
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar
```

```python
# Create datastory
metadata = {
    "name": "EChart Calendar examples",
    "keywords": ["calendar","chart","examples"]
}

ds = deetly.story(metadata)
```

```python
begin = datetime.date(2019, 1, 1)
end = datetime.date(2019, 12, 31)
data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
    for i in range((end - begin).days + 1)
]

g = (
    Calendar()
    .add("", data, calendar_opts=opts.CalendarOpts(range_="2019"))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            max_=25000,
            min_=500,
            orient="horizontal",
            is_piecewise=True,
            pos_top="230px",
            pos_left="100px",
        )
    )
)

g.render_notebook()
```

```python
ds.echart(g, "Calendar")
```

```python
ds.publish()
```
