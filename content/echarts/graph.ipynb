```python
import deetly
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Graph
```

```python
# Create datastory
metadata = {
    "name": "EChart Graph examples",
    "keywords": ["graph","chart","examples"]
}

ds = deetly.story(metadata)
```

```python
nodes = [
    {"name": "A", "symbolSize": 10},
    {"name": "B", "symbolSize": 20},
    {"name": "C", "symbolSize": 30},
    {"name": "D", "symbolSize": 40},
]

links = []
for i in nodes:
    for j in nodes:
        links.append({"source": i.get("name"), "target": j.get("name")})

g = Graph().add("", nodes, links, repulsion=8000)
g.render_notebook()
```

```python
ds.echart(g, "Basic Graph")
```

```python
g = Graph().add(
        "", nodes, links, repulsion=5000, edge_label=opts.LabelOpts(is_show=True)
)
g.render_notebook()
```

```python
ds.echart(g, "Graph with edges")
```

```python
g = Graph().add("", nodes, links, repulsion=8000, is_draggable=True, symbol_size=50)
g.render_notebook()
```

```python
ds.echart(g, "Graph with draggable nodes")
```

```python
ds.publish()
```
