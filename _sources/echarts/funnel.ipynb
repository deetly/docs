# Funnel

>Sample funnel charts

Source: pyecharts test

```python
import deetly
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Funnel
```

```python
# Create datastory
metadata = {
    "name": "EChart Funnel examples",
    "keywords": ["funnel","chart","examples"]
}

ds = deetly.story(metadata)
```

```python
items = [
('1', 50),
('2', 22),
('3', 34),
('4', 68),
('5', 20),
('6', 105),
('7', 53)]


funnel_item = [opts.FunnelItem(name=d[0], value=d[1]) for d in items]

g = (
    Funnel()
    .add("Item", funnel_item)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Funnel"), 
        visualmap_opts=opts.VisualMapOpts(pos_top="230px", pos_left="100px")
    )
)

g.render_notebook()
```

```python
ds.echart(g, "Funnel")
```

```python
ds.publish()
```
