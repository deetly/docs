# Gauge

>Sample gauge charts

Source: pyecharts test

```python
import deetly
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Gauge
```

```python
# Create datastory
metadata = {
    "name": "EChart Gauge examples",
    "keywords": ["gauge","chart","examples"]
}

ds = deetly.story(metadata)
```

```python
g = Gauge().add("", [("", 66.6)])

g.render_notebook()
```

```python
ds.echart(g, "Basic Gauge")
```

```python
g = Gauge().add("",[("", 66.6)],radius="50%")

g.render_notebook()
```

```python
ds.echart(g, "Small Gauge")
```

```python
ds.publish()
```
