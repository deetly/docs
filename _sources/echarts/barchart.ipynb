# Bar chart

>Basic bar chart

```python
import deetly
import pandas as pd

from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ChartType
```

```python
# Create datastory
metadata = {
    "name": "EChart Barchart examples",
    "keywords": ["bar","barchart","examples"]
}

ds = deetly.story(metadata)
```

```python
g = (Bar()
    .add_xaxis(["Tokyo", "Delhi", "Shanghai", "SÃ£o Paulo", "Mexico City", "Cairo"])
    .add_yaxis('Population', [37.4, 28.5, 25.6, 21.6, 21.6, 20.1])
    .set_global_opts(title_opts=opts.TitleOpts(title="Worlds largest cities", subtitle="UN 2018 estimates. Million"))
)

g.render_notebook()
```

```python
ds.echart(g, "World cities")
```

```python
g = (Bar()
    .add_xaxis(["A", "B", "C"])
    .add_yaxis("series0", [1, 2, 4])
    .add_yaxis("series1", [2, 3, 6])
    .add_yaxis("series2", [5, 8, 7])
    .set_series_opts(stack="stack")
)
g.render_notebook()
```

```python
ds.echart(g, "Stacked")
```

```python
g = (Bar()
    .add_xaxis(["A", "B", "C"])
    .add_yaxis("series0", [1, 2, 4])
    .set_global_opts(
        graphic_opts=[
            opts.GraphicImage(
                graphic_item=opts.GraphicItem(
                    id_="logo",
                    left=20,
                    bottom=20,
                    z=10,
                    bounding="raw",
                    origin=[75, 75],
                ),
                graphic_imagestyle_opts=opts.GraphicImageStyleOpts(
                    image="https://public.deetly.com/favicon.ico",
                    width=20,
                    height=20,
                    opacity=1,
                ),
            )
        ]
    )
)
g.render_notebook()
```

```python
ds.echart(g, "With logo")
```

```python
from pyecharts.commons.utils import JsCode

g = (
    Bar(
        init_opts=opts.InitOpts(
            bg_color={
                "type": "pattern",
                "image": JsCode("img"),
                "repeat": "no-repeat",
            }
        )
    )
    .add_xaxis(["A", "B", "C"])
    .add_yaxis("series0", [1, 2, 4])
    .add_yaxis("series1", [2, 3, 6])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="Bar",
            subtitle="with bg",
            title_textstyle_opts=opts.TextStyleOpts(color="white"),
        )
    )
)

g.add_js_funcs(
    """
    var img = new Image(); img.src = 'https://raw.githubusercontent.com/deetly/docs/master/images/gaspar-manuel-zaldo-5YZU84pvH8Y-unsplash.jpg?token=AAVTYSSZY44Z2AIEB62JUCDAJDFB6';
    """
)

g.render_notebook()
```
