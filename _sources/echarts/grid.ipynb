# Grid chart

>Line + Bar in Grid

```python
import deetly
import pandas as pd

from pyecharts.charts import Bar, Line, Grid
from pyecharts import options as opts
from pyecharts.globals import ChartType
```

```python
# Create datapackage
description = """
ECharts Grid examples
"""

metadata = {
    "name": "EChart Grid examples",
    "description":description, 
    "author": "deetly",
    "keywords": ["grid","echarts", "chart","examples"],
    "theme":"examples",
    "license": "MIT"
}

dp = deetly.package(metadata)
```

```python
line = (Line()
    .add_xaxis(["Tokyo", "Delhi", "Shanghai", "São Paulo", "Mexico City", "Cairo"])
    .add_yaxis('Population', [37.4, 28.5, 25.6, 21.6, 21.6, 20.1])
    .add_yaxis('Growth', [7.4, 8.5, 5.6, 1.6, 1.6, 2.1])
)

bar = (Bar()
    .add_xaxis(["Tokyo", "Delhi", "Shanghai", "São Paulo", "Mexico City", "Cairo"])
    .add_yaxis('Population', [37.4, 28.5, 25.6, 21.6, 21.6, 20.1])
    .add_yaxis('Growth', [7.4, 8.5, 5.6, 1.6, 1.6, 2.1])
    .set_global_opts(title_opts=opts.TitleOpts(title="Worlds largest cities", subtitle="UN 2018 estimates. Million"))
)

g = (Grid()
    .add(bar, opts.GridOpts(pos_left="5%", width="60%"))
    .add(line, opts.GridOpts(pos_left="70%"))
)

g.render_notebook()
```

```python
# add chart to package
dp.echart(g, "Population", "Worlds largest cities")
```

```python
def _chart_for_grid() -> Bar:
    x_data = ["{}".format(i) for i in range(1, 13)]
    bar = (
        Bar()
        .add_xaxis(x_data)
        .add_yaxis(
            "2019",
            [2.0, 4.9, 7.0, 23.2, 25.6, 76.7, 135.6, 162.2, 32.6, 20.0, 6.4, 3.3],
            yaxis_index=0,
        )
        .add_yaxis(
            "2020",
            [2.6, 5.9, 9.0, 26.4, 28.7, 70.7, 175.6, 182.2, 48.7, 18.8, 6.0, 2.3],
            yaxis_index=1,
        )
        .set_series_opts(stack="MY_STACK_NAME")
        .extend_axis(yaxis=opts.AxisOpts(name="2020", type_="value", position="right"))
        .extend_axis(yaxis=opts.AxisOpts(type_="value", name="Change", position="left"))
        .set_global_opts(
            yaxis_opts=opts.AxisOpts(name="2019", position="right", offset=80)
        )
    )

    line = (
        Line()
        .add_xaxis(x_data)
        .add_yaxis(
            "Change",
            [2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2],
            yaxis_index=2,
            color="#675bba",
            label_opts=opts.LabelOpts(is_show=False),
        )
    )

    bar.overlap(line)
    return bar
```

```python
bar = _chart_for_grid()
g = Grid().add(
    bar, 
    opts.GridOpts(pos_left="5%", pos_right="20%", 
    is_contain_label=True), 
    is_control_axis_index=True
)
g.render_notebook()
```

```python
dp.echart(g, "Growth", "Growth")
```
