```python
import deetly

from pyecharts.charts import Boxplot
```

```python
# Create datastory
metadata = {
    "name": "EChart Boxplot examples",
    "keywords": ["boxplot","chart","examples"]
}

ds = deetly.story(metadata)
```

```python
v1 = [
    [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880, 1000, 980],
    [960, 940, 960, 940, 880, 800, 850, 880, 900, 840, 830, 790],
]
v2 = [
    [890, 810, 810, 820, 800, 770, 760, 740, 750, 760, 910, 920],
    [890, 840, 780, 810, 760, 810, 790, 810, 820, 850, 870, 870],
]

g = Boxplot()
g.add_xaxis(["Experiment 1", "Experiment 2"])
g.add_yaxis("A", g.prepare_data(v1))
g.add_yaxis("B", g.prepare_data(v2))

g.render_notebook()
```

```python
ds.echart(g, "Boxplot")
```

```python
ds.publish()
```
