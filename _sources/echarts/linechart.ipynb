# Line chart

>Basic line chart

```python
import deetly
import pandas as pd

from pyecharts.charts import Line
from pyecharts import options as opts
from pyecharts.globals import ChartType
from pyecharts.options import LineStyleOpts
```

```python
# Create datapackage
description = """
ECharts Linechart examples
"""

metadata = {
    "name": "EChart Linechart examples",
    "description":description, 
    "author": "deetly",
    "keywords": ["line","linechart","chart","examples"],
    "theme":"examples",
    "license": "MIT"
}

dp = deetly.package(metadata)
```

```python
g = (Line()
    .add_xaxis(["Tokyo", "Delhi", "Shanghai", "São Paulo", "Mexico City", "Cairo"])
    .add_yaxis('Population', [37.4, 28.5, 25.6, 21.6, 21.6, 20.1])
    .set_global_opts(title_opts=opts.TitleOpts(title="Worlds largest cities", subtitle="UN 2018 estimates. Million"))
)

g.render_notebook()
```

```python
# add chart to package
dp.echart(g, "Population", "Worlds largest cities")
```

```python
g = (Line()
    .add_xaxis(["Tokyo", "Delhi", "Shanghai", "São Paulo", "Mexico City", "Cairo"])
    .add_yaxis('Population', [37.4, 28.5, 25.6, 21.6, 21.6, 20.1])
    .set_series_opts(linestyle_opts=LineStyleOpts(curve=50))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Worlds largest cities", subtitle="UN 2018 estimates. Million"),
        xaxis_opts=opts.AxisOpts(is_inverse=True),
        #datazoom_opts=opts.DataZoomOpts(type_="value"),
        visualmap_opts=opts.VisualMapOpts(type_="size", range_size=[0, 100]),
    )
)

g.render_notebook()
```

```python
# publish package
dp.publish()
```
